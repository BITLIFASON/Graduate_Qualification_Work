```python
def preprocess_text(text):
    tokens = text.lower().strip().split(' ')
    for token in tokens:
        token = token.replace('\n','')
        token = re.sub(r'\d+[\,\.]?\d+','',token)
        token = re.sub(r'[\.\,\"\'\;\:\)\]\(\[\?\!\-\+]{1,}$','',token)
        token = re.sub(r'^[\.\,\(\[\"\'\;\:\)\]\?\!\-\+]{1,}','',token)
    if tokens != "":
        return ' '.join(tokens)
```

```python
with open('ohsumed_dataset.pkl','rb') as ohsumed_file:
     source = pickle.load(ohsumed_file)
```

```python
%%time
source['texts'] = source.texts.apply(preprocess_text).values
```

```python
target_column = 'Y'
```

```python
X = source.texts.to_list()
```

```python
Y = source[target_column].to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=RANDOM_SEED,stratify=source[target_column].to_list())
```
